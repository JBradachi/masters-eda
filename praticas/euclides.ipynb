# Pratica MDC

## Introdução

O máximo divisor comum é um problema clássico da computação, a partir dele podemos observar o quantoa escolha dos algorítmos influenciam na eficiencia e no tempo de resposta.

Vamos observar duas implementações naive e uma do algorítimo de euclides

```python
import time
from typing import Callable
```

```python
dados = [
    (15, 5),
    (20, 12),
    (100, 50),
    (20002093920, 2039920),
    (1280937949032, 21930830),
    (32180957200000, 18200000)
]

def testes(funcao:Callable[[int, int], int], dados:list):
    
    for dado in dados:
        tempo = time.time()
        resultado = funcao(dado[0], dado[1])
        tempo =  time.time() - tempo
        print(f"O MMC de {dado[0]} e {dado[1]} é {resultado}, tempo: {tempo}")
```

```python
def gcd1(m:int, n:int):
    if 0 in [m,n]:
        return max(m,n)
    maximo = 1
    for i in range(1, min(m, n)+1):
        if m%i == 0 and n%i == 0:
            maximo = i
    return maximo

def gcd2(m:int, n:int):
    if 0 in [m,n]:
        return max(m,n)

    for i in range(min(m, n), 0, -1):
        if m%i == 0 and n%i == 0:
            return i

```

```python
testes(gcd1, dados)
```

```python
testes(gcd2, dados)
```

Observe o quanto o tempo escala conforme o crescimento dos valores, mas com um bom algorítmo, como o de euclides, o custo computacional se estabiliza

```python
def euclides_recursivo(m, n):
    if n == 0: return m
    return euclides_recursivo(n, m%n)

def euclides(m,n):
    i=True
    while i:
        if n == 0:
            i = False 
            return m
        a = m
        m = n
        n = a%n
```

```python
testes(euclides_recursivo, dados)
```

```python
testes(euclides, dados)
```
